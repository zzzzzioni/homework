## if-else문

1. if문은 조건이 참(True)일 때 그 아래 들여쓰기 된 코드를 실행합니다.


2. if-else문은 if문의 조건문이 참(True)일 때는 if문에 해당하는 코드를 실행하고,

   참이 아니라면 else문에 해당하는 코드를 실행합니다.
   
   
3. if-elif-else문은 if문의 조건문의 참,거짓 여부를 판단하고 if문이 참이라면 if문에 해당하는 코드를 실행합니다.

   if문이 참이 아니라면 elif문의 조건문의 참, 거짓 여부를 판단하고 참이라면 해당하는 코드를 실행합니다.
   
   모든 조건문이 참이 아니라면 else문에 해당하는 코드를 실행합니다.
   
   
```python3
if 조건:
    실행할 코드
    
if 조건:
    실행할 코드
else:
    실행할 코드
    
    
if 조건1:
    실행할 코드
elif 조건2:
    실행할 코드
else:
    실행할 코드
```

````python
# if문은 조건문이 True라면 실행됩니다.
money = True

if money:
    print("money")
````

````python
money = 10000

if money > 5000:
    print("enough money")
else:
    print("no money")
````

````python
# 가진 돈에 따라서 점심메뉴를 나눈다.
# 4000원 미만은 '학식', 4000원 이상 8000원 미만은 '맥도날드', 8000원 이상은 '스테이크'를 print.

money = 5000

if money < 4000:
    print('학식')
elif money < 8000:
    print('맥도날드')
else:
    print('스테이크')
````

````python
## 순서대로 조건문을 판단하기 때문에 원하는 결과가 안나올 수 있으므로 주의해주세요!!

money = 5000

if money < 8000:
    print('맥도날드')
elif money < 4000:
    print('학식')
else:
    print('스테이크')
````

3. '간격'은 생략할 때가 많으며, 생략할 경우 1로 지정됩니다.


4. '시작'과 '간격'을 모두 생략하고 '끝'만 지정해주면 '시작'은 0, '간격'은 1로 지정됩니다.

````python
print(len(numbers))
for i in range(len(numbers)):
    print(numbers[i])
````

````python
for i in range(5):
    print(i)
````

````python
for i in range(0, 10, 2):
    print(i)
````

````python
i = 0
for name in ['주현경', '김희연', '금종연'] :
    print(i, name)
    i += 1
````

````python
dict = {'학생A' : '컴과', '학생B': '산공', '학생C' : '둘다'}

for key, val in dict.items() :
    print(f'이름 : {key}, 학과 : {val}')
````

3. if문과 continue를 사용해서 while문의 처음으로 돌아갑니다

````python
count = 0

while count < 10:
    count += 1
    print(count)
````

````python
count = 0

while True:
    count += 1
    print(count)
    if count == 10:
        break
````

````python
count = 0

while count < 10:
    count += 1
    if count % 2 == 0:
        continue
    print(count)
````

````python
def add(num1, num2):
    return num1 + num2

a = 1
b = 2
c = add(a, b)
print(c)
````

````python
# argument는 position 기반 또는 파라미터에 값을 직접 넘겨줄 수 있습니다.
def divide(num1, num2):
    return num1 / num2

a = 10
b = 20

relative_1 = divide(a, b)
relative_2 = divide(b, a)
absolute_1 = divide(num1=a, num2=b)
absolute_2 = divide(num2=a, num1=b)

print('relative_1:', relative_1)
print('relative_2:', relative_2)
print('absolute_1:', absolute_1)
print('absolute_2:', absolute_2)
````

````python
# 전역변수와 지역변수를 구분해야합니다.
x = 10

def print_x():
    print(x)

print_x()
````

````python
x = 10
def modify_x():
    x = 100
    print(x)

modify_x()
print(x)
````

````python
def modify_lst(lst):
    for i in range(len(lst)):
        lst[i] += 10
    print(lst)

lst = [1, 2, 3, 4, 5]
modify_lst(lst)
print(lst)
````

## 람다 표현식 (lambda expression)

지금까지 def로 함수를 만드는 방법을 알아봤습니다


람다 표현식은 익명 함수를 만들 수 있는 기능입니다.

<pre>
<code>
lambda 변수: 식
</code>
</pre>

람다 표현식 안에서는 변수를 만들 수 없습니다.

````python
def multiply_ten(num):
    return num * 10

num = 10
print(multiply_ten(num))
````

````python
# lambda 매개변수 : 표현식
a = (lambda x: x * 10)(2)
a
````

````python
y = 10
(lambda x: x + y)(1)
````

````python
# map 함수는, 두 번째 인자로 들어온 반복 가능한 자료형 (리스트나 튜플)을, 첫 번째 인자로 들어온 함수에 하나씩 집어넣어서 함수를 수행
## 파이썬에서는 함수도 객체
print(list(map(multiply_ten, [10, 20, 30])))
print(list(map(lambda x: x * 10, [10, 20, 30])))

# sorted 함수는 sorted(리스트 또는 튜플, 함수)
## dict 정렬 시 key를 정렬하게 된다
d = {'a': 100, 'b': 50, 'c': 10}
print(sorted(d))
print(sorted(d, key=lambda x: d[x])) # sort by number not alphabet
````

## 클래스와 객체

클래스란 객체를 정의해놓은 것 또는 객체의 설계도, 틀이라고 할 수 있습니다.


클래스는 객체를 생성하는데 사용되며, 객체는 클래스에 정의된 대로 생성됩니다.


프로그래밍에서 객체는 클래스에 정의된 내용대로 메모리에 생성된 것을 말합니다.


A 클래스로 만들어진 a 라는 객체가 있다면 a를 A 클래스의 인스턴스(instance)라고 부릅니다.

*클래스가 왜 필요? 객체 지향?*
- 코드의 재사용과 중복을 방지하기 위해서
- 함수 중심의 프로그래밍은 데이터가 너무 방대해지기 때문에 보수 및 수정이 어려워진다.
- 클래스를 중심으로 코딩? -> 데이터 중심의 프로그래밍 -> 객체로 코드가 관리되기 때문에 유지 보수가 쉬움.


가장 간단한 클래스의 모습은 다음과 같습니다.
<pre>
<code>
class 클래스명:
    pass
</code>
</pre>


클래스에는 변수, 생성자, 메소드 등을 정의 할 수 있습니다.


아래의 코드에서 self는 자기 자신을 뜻합니다.

<pre>
<code>
(변수 설정)
class 클래스명:
    self.a = 0
    self.b = 4
</code>
</pre>

<pre>
<code>
(생성자)
class 클래스명:
    def __init__(self, 변수1, 변수2, ...):
        self.변수명1 = 변수1
        self.변수명2 = 변수2
</code>
</pre>
<pre>
<code>
(메소드)
class 클래스명:
    def add(self, 변수1, 변수2):
        print(변수1 + 변수2)
</code>
</pre>

**튜플**을 오름차순 정렬해서 리스트로 반환하는 함수를 만들어주세요.
```sorted()``` 함수는 금지입니다. 버블 정렬, 삽입 정렬, 퀵 정렬 등 원하는 정렬 알고리즘을 선택해주세요


```python3
tup = (3, 4, 1)
lst = sort(tup)
# [1, 3, 4]가 나와야 함
print(lst)
```

````python
def sort(tup):
    # 이 부분은 지우지 말아주세요
    if not isinstance(tup, tuple):
        raise TypeError("tuple을 넣어주세요")
    length = len(tup)
    tup_len_list = list(range(length))
    for i in range(length):
        a = 0
        for j in range(length):
            if tup[i] > tup[j] :
                a+=1
            if tup[i] <= tup[j] :
                pass
        tup_len_list[a] = tup[i]
    return tup_len_list
````

## 과제 2

1. 'Point' 라는 이름의 클래스를 만들어 주세요.

2. 'Point' 클래스는 0이 아닌 정수 x와 y를 인스턴스 변수를 생성자에 입력 받습니다.

3. 'Point' 클래스는 위치를 set하는 함수 ```set_position(x, y)```이 있습니다.

4. 'Point' 클래스는 위치를 get하는 함수 ```get_position()```이 있습니다.

5. 'Point' 클래스는 다른 Point instance와의 거리를 계산하는 ```get_dist()``` 함수가 있습니다.

6. 'Point' 클래스는 ```print(Point)```를 했을 때 Point(x: x, y: y)를 출력해야합니다.

````python
from math import sqrt

class Point:
    # pass를 지우고 클래스를 완성해주세요!!
    def __init__(self, x, y):
        self.x = x
        self.y = y
        
    def set_point(self, x, y):
        self.x = x
        self.y = y
    
    def get_point(self):
        print('x:', str(self.x), 'y:', str(self.y))
    
    def get_dist(p1, p2):
        return sqrt( (p1.x - p2.x)**2 + (p1.y - p2.y)**2 )
        
    def __str__(self):
        return 'Point(x: {}, y: {})'.format(self.x, self.y)
        
        
````

````python
# 출력
# x: 1, y: 2
p1 = Point(1, 2)
p1.get_point()

# 출력
# x: 10, y: 20
p1.set_point(10,20)
p1.get_point()

# (10, 20)과 (3, 4)의 유클리디안 거리를 계산해주세요
p2 = Point(3, 4)
print(p1.get_dist(p2))

print(p1)
print(p2)
````
